```python

def getDic(input_path):
    Sequences = {}
    amountSeq = 0
    counter = 1
    lenOfConent = 0
    with open(input_path, 'r') as file:
        content = file.read()

    split = content.split('\n')
    amountSeq = int(split[0].split(' ')[1])
    Timeout_length = int(split[0].split(' ')[2])
    lenOfConent = len(split)
    for x in range(1, amountSeq+1):
        name = split[counter].split(" ")[0]
        sequence = split[counter].replace(" ", "").replace(name, "")
        Sequences[name] = sequence
        for x in range(counter + amountSeq + 1,lenOfConent,amountSeq + 1):
            Sequences[name] = Sequences[name] + split[x].replace(" ", "")
        counter=counter+1
    return Sequences
```

```python
def combine_alignments(mainfile, fileToAdd):
    newDic = {}
    val = random.choice(list(fileToAdd.items()))
    seq_len = len(val[1])

    for x in mainfile.keys():
        if x in fileToAdd.keys():
            newDic[x] = mainfile[x] + fileToAdd[x]
        else:
            newDic[x] = mainfile[x] + "-"*seq_len
    return newDic
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

input_Timeout = '/Users/julius/Gobi/Final_results/Mafft_alignmet/Mafft_alignment_linsi_maxiterate1000_localpair.phy'
input_Timeless = '/Users/julius/Gobi/Final_results/Mafft_alignmet/Timeless_Mafft_alignment_linsi_maxiterate1000_localpair.phy'

def dict_to_phy(dictionary, filename):
    # Create a list to hold SeqRecord objects
    seq_records = []
    
    for species, sequence in dictionary.items():
        # Create a SeqRecord object for each entry in the dictionary
        seq_record = SeqRecord(Seq(sequence),
                               id=species,  # Use the species name as the ID
                               description="")  # No description
        seq_records.append(seq_record)
    
    # Write the sequence records to a Phylip file
    SeqIO.write(seq_records, filename, "phylip")


alignments = combine_alignments(getDic(input_Timeout), getDic(input_Timeless))

dict_to_phy(alignments, '/Users/julius/Gobi/Final_results/Mafft_alignmet/combined_alignment.phy')
```
